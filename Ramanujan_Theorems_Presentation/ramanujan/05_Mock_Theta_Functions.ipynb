# Ramanujan’s Mock Theta Functions

In the final months of his life (1919–1920), Ramanujan wrote letters to G. H. Hardy describing a new type of function he called **"Mock Theta Functions"**.  

He listed 17 examples, but without formal definitions.  
These mysterious functions seemed similar to the classical theta functions of modular form theory — yet they did **not** behave exactly like modular forms.  

For decades, mathematicians puzzled over what “mock” meant, until **Sander Zwegers (2002)** showed that Ramanujan’s mock theta functions are related to **harmonic Maass forms** — completing a century-old mystery.

---

### A Simple Example

One of Ramanujan’s third-order mock theta functions is:

$$
f(q) = 1 + \sum_{n=1}^{\infty} 
\frac{q^{n^2}}{(1+q)^2 (1+q^2)^2 \cdots (1+q^n)^2}
$$

We can approximate \( f(q) \) numerically for \( |q| < 1 \) and explore how it behaves as \( q \to 1^- \).


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mock_theta_f(q, n_terms=50):
    """
    Compute a truncated approximation of Ramanujan’s third-order mock theta function f(q).
    f(q) = 1 + sum_{n=1}^∞ q^{n^2} / ∏_{k=1}^{n} (1 + q^k)^2
    """
    total = 1.0
    for n in range(1, n_terms + 1):
        numerator = q**(n**2)
        denom = np.prod([(1 + q**k)**2 for k in range(1, n + 1)])
        total += numerator / denom
    return total
```

```python
q_values = np.linspace(0.1, 0.99, 50)
f_values = np.array([mock_theta_f(q, n_terms=50) for q in q_values])

plt.figure(figsize=(9,6))
plt.plot(q_values, np.real(f_values), label="Re[f(q)]", linewidth=2)
plt.plot(q_values, np.imag(f_values), '--', label="Im[f(q)]", linewidth=2)
plt.title("Ramanujan’s Third-Order Mock Theta Function f(q)")
plt.xlabel("q")
plt.ylabel("f(q)")
plt.grid(True, alpha=0.6)
plt.legend()
plt.show()
```

### Mathematical Significance

Ramanujan’s mock theta functions are now understood as **holomorphic parts** of **harmonic Maass forms** — functions that generalize modular forms by allowing a non-holomorphic “shadow” term.  

In modern notation, a mock theta function \( f(q) \) is paired with a modular form \( g(q) \) such that:

$$
\widehat{f}(q) = f(q) + g^*(q)
$$

where \( g^*(q) \) is a non-holomorphic integral transform of \( g(q) \).

These discoveries revealed deep connections between:
- **Number theory**,  
- **Modular forms**,  
- **String theory**, and  
- **Black hole entropy** (via modern physics interpretations of modularity).

---

Ramanujan anticipated — without formal proof — structures that wouldn’t be fully understood until nearly a **century later**.  
This is perhaps the clearest example of his mathematical intuition operating far ahead of its time.


```python
plt.figure(figsize=(9,6))
plt.plot(q_values, np.abs(f_values), 'm-', linewidth=2)
plt.title("Magnitude |f(q)| of Ramanujan’s Mock Theta Function")
plt.xlabel("q")
plt.ylabel(r"$|f(q)|$")
plt.grid(True, alpha=0.6)
plt.show()
```

### Discussion

Ramanujan’s Mock Theta Functions represent the culmination of his mathematical journey.

- They **bridge** the gap between classical modular forms and modern number theory.  
- Their structure inspired major breakthroughs in the **Langlands program**, **string theory**, and **quantum modular forms**.  
- The rediscovery of their meaning by Zwegers (2002) and subsequent work by Ono and Bringmann confirmed Ramanujan’s **deep foresight**.
